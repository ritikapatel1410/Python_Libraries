 @Title : display a bar chart of the popularity of programming Languages.Use different color for each bar.

```python
import matplotlib.pyplot as plt
```

```python
Programming_languages=["Java", "Python", "PHP", "JavaScript", "C#", "C++"]
Popularity=[22.2, 17.6, 8.8, 8, 7.7, 6.7]
colour_list=["green","blue","black","red","purple","yellow"]
```

```python
plt.bar(Programming_languages,Popularity,color=colour_list)
plt.xlabel("Programming_languages")
plt.ylabel("Popularity")
plt.title("popularity of programming Languages")
plt.show()
```
