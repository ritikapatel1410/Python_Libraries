 @Title : Write a Python program to create a stacked bar plot with error bars
          Note: Use bottom to stack the women?s bars on top of the men?s bars.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
menMeans = (22, 30, 35, 35, 26)
womenMeans = (25, 32, 30, 35, 29)
menStd = (4, 3, 4, 1, 5)
womenStd = (3, 5, 2, 3, 3)
index=np.arange(len(menMeans))
```

```python
Bar1=plt.bar(index,menMeans,width=0.5, yerr=menStd, color='red')
Bar2=plt.bar(index,womenMeans,width=0.5,bottom=menMeans, yerr=womenStd, color='green')
plt.xlabel("Group")
plt.ylabel("Score")
plt.xticks(index,('Group1', 'Group2', 'Group3', 'Group4', 'Group5'))
plt.yticks(np.arange(0, 81, 10))
plt.title('Scores by group\n' + 'and gender')
plt.legend((Bar1[0], Bar2[0]), ('Men', 'Women'))
plt.show()
```
