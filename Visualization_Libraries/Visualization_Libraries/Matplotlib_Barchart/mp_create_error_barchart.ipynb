 @Title : Python program to create bar plots with error bars on the same figure

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
menMeans = (54.74, 42.35, 67.37, 58.24, 30.25)
menStd = (4, 3, 4, 1, 5)
index = np.arange(5)
```

```python
plt.bar(index,menMeans,width=0.5, yerr=menStd, color='red')
plt.xlabel("Velocity")
plt.ylabel("Score")
plt.title("Scores by Velocity")
plt.show()
```
