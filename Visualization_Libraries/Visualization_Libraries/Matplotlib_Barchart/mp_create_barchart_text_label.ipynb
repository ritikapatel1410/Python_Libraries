 @Title : display a bar chart of the popularity of programming Languages. Attach a text label above each bar                     displaying its popularity (float value).

```python
import matplotlib.pyplot as plt
```

```python
Programming_languages=["Java", "Python", "PHP", "JavaScript", "C#", "C++"]
Popularity=[22.2, 17.6, 8.8, 8, 7.7, 6.7]
```

```python
fig, ax = plt.subplots()
rects1 = ax.bar(Programming_languages, Popularity)
plt.bar(Programming_languages,Popularity,color="b")
plt.xlabel("Programming_languages")
plt.ylabel("Popularity")
plt.title("popularity of programming Languages")
def autolabel(rects):
    """
    Attach a text label above each bar displaying its height
    """
    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()/2., 1.05*height,
                '%0.3f' % float(height),
        ha='center', va='bottom')      
autolabel(rects1)
plt.show()
```
