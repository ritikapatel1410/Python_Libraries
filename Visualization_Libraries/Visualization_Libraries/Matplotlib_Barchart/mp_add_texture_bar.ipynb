 @Title : add textures (black and white) to bars and wedges

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure()
patterns = [ "|" , "\\" , "/" , "+" , "-", ".", "*","x", "o", "O" ]
ax = fig.add_subplot(111)
for index_value in range(len(patterns)):
    ax.bar(index_value, 3, color='white', edgecolor='black', hatch=patterns[index_value])
plt.show()
```
