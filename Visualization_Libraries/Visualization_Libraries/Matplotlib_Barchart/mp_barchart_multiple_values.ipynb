 @Title : create bar plot of scores by group and gender. Use multiple X values on the same chart for men and women.

```python
import matplotlib.pyplot as plt
import numpy as np 
```

```python
n_groups = 5
men_means = (22, 30, 33, 30, 26)
women_means = (25, 32, 30, 35, 29)
width = 0.27 
```

```python
index=np.arange(n_groups)
fig, ax = plt.subplots()
ax.bar(index, men_means, width, color='b', label='men_means')
ax.bar(index+width, women_means, width, color='g', label='women_means')
ax.set_ylabel('Score')
ax.set_xlabel('Person')
plt.legend()
ax.set_xticks(index+width)
ax.set_xticklabels( ('G1', 'G2', 'G3', 'G4', 'G5') )
plt.title("chart for men and women")
plt.show()
```
