 @Title : display a bar chart of the popularity of programming Languages.Specify the position and width of each bar              plot

```python
import matplotlib.pyplot as plt
```

```python
Programming_languages=["Java", "Python", "PHP", "JavaScript", "C#", "C++"]
Popularity=[22.2, 17.6, 8.8, 8, 7.7, 6.7]
```

```python
y_pos = [0,1,4,7,9,10]
width=[0.1,0.2,0.5,1.1,0.2,0.3]
plt.bar(y_pos,Popularity,width)
plt.xlabel("Programming_languages")
plt.ylabel("Popularity")
plt.title("popularity of programming Languages")
plt.show()
```
