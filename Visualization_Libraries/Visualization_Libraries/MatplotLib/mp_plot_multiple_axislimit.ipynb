 @Title : display the current axis limits values and set new axis values

```python
import matplotlib.pyplot as plt 
import numpy as np
```

```python
y_axis_values_1=np.array([0,1,2,3,10])
y_axis_values_2=np.array([3,0,1,7,10])
```

```python
plt.plot(y_axis_values_1,color="b",label="Line 1",marker="*",mfc='red', markersize=12,mec="black")
plt.plot(y_axis_values_2,color="g",label="Line2",marker="o",mfc='red', markersize=12,mec="black")
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.legend()
plt.title("display the current axis limits values")
plt.show()
```

```python
plt.plot(y_axis_values_1,color="b",label="Line 1",marker="*",mfc='red', markersize=12,mec="black")
plt.plot(y_axis_values_2,color="g",label="Line2",marker="o",mfc='red', markersize=12,mec="black")
plt.ylim(0,8)
plt.xlim(0,8)
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.legend()
plt.title("set new axis values")
plt.show()
```
