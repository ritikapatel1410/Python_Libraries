 @Title : plot two or more lines and set the line markers

```python
import matplotlib.pyplot as plt 
import numpy as np
```

```python
y_axis_values_1=np.array([0,1,2,3])
y_axis_values_2=np.array([3,0,1,7])
```

```python
plt.plot(y_axis_values_1,color="b",label="Line 1",marker="*",mfc='red', markersize=12,mec="black")
plt.plot(y_axis_values_2,color="g",label="Line2",marker="o",mfc='red', markersize=12,mec="black")
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.legend()
plt.title("Two or more lines on same plot with suitable legends")
plt.show()
```
