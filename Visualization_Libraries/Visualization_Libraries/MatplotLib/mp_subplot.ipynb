```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
plot_one_x_axis=np.array([1,2,3])
plot_one_y_axis=np.array([1,4,5])
plot_two_x_axis=np.array([2,5,6])
plot_two_y_axis=np.array([12,6,5])
```

```python
plt.subplot(1,2,1)
plt.plot(plot_one_x_axis,plot_one_y_axis,linestyle=":",c="g")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("first plot")
plt.subplot(1,2,2)
plt.plot(plot_two_x_axis,plot_two_y_axis,linestyle=":",c="g")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("second plot")
plt.show()
```
