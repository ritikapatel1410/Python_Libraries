 @Title : plot quantities which have an x and y position

```python
import matplotlib.pyplot as plt 
```

```python
x_axis_values1 = [2, 3, 5, 6, 8]
y_axis_values1 = [1, 5, 10, 18, 20]
x_axis_values2 = [3, 4, 6, 7, 9]
y_axis_values2 = [2, 6, 11, 20, 22]
```

```python
plt.axis([0, 10, 0, 30])
plt.plot(x_axis_values1,y_axis_values1,'b*',x_axis_values2,y_axis_values2,'ro')
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.title("plot quantities")
plt.show()
```
