 @Title : plot two or more lines with different styles

```python
import matplotlib.pyplot as plt 
import numpy as np
```

```python
y_axis_values_1=np.array([0,1,2,3])
y_axis_values_2=np.array([3,4,5,6])
```

```python
plt.plot(y_axis_values_1,color="b",linewidth=3,label="Line with width 3",linestyle=":")
plt.plot(y_axis_values_2,color="g",linewidth=5,label="Line with width 5",linestyle="-.")
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.legend()
plt.title("Two or more lines on same plot with suitable legends")
plt.show()
```
