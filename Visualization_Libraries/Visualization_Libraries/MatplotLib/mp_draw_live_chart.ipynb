 @Title : draw line charts of the financial data

```python
import matplotlib.pyplot as plt 
import pandas as pd
```

```python
dataframe = pd.read_csv('csv_file/data.csv')
Date=dataframe["Date"]
Open=dataframe["Open"]
High=dataframe["High"]
Low=dataframe["Low"]
Close=dataframe["Close"]

plt.plot(Date,Open,label="Open",color="red")
plt.plot(Date,High,label="High",color="black")
plt.plot(Date,Low,label="Low",color="green")
plt.plot(Date,Close,label="Close",color="yellow")
plt.legend()
plt.xlabel("Date")
plt.ylabel("Information")
plt.title("charts of the financial data")
plt.show()
```
