 @Title : plot several lines with different format styles in one command using arrays.

```python
import matplotlib.pyplot as plt
import numpy as np 
```

```python
values_range=np.arange(0.,5.,0.2)
```

```python
plt.plot(values_range, values_range, 'g--', values_range, values_range**2, 'bs', values_range, values_range**3, 'r^')
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("plot several lines with different format styles")
plt.show()
```
