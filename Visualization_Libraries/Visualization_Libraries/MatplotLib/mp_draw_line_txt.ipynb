 @Title : draw a line using given axis values taken from a text file, with suitable label in the x axis, y axis and a title

```python
import matplotlib.pyplot as plt 
import numpy as np
```

```python
with open("test/test.txt") as file:
    data=file.read()
data=data.split("\n")
x_axis_values=[x_axis.split(' ')[0] for x_axis in data]
y_axis_values=[x_axis.split(' ')[1] for x_axis in data]
```

```python
plt.plot(x_axis_values,y_axis_values)
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.title("Draw line")
```

```python
plt.show()
```
