 @Title : plot two or more lines on same plot with suitable legends of each line

```python
import matplotlib.pyplot as plt 
import numpy as np
```

```python
y_axis_values_1=np.array([0,1,2,3])
y_axis_values_2=np.array([3,4,5,6])
```

```python
plt.plot(y_axis_values_1,label="Line1")
plt.plot(y_axis_values_2,label="Line2")
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.legend()
plt.title("Two or more lines on same plot with suitable legends")

```

```python
plt.show()
```
