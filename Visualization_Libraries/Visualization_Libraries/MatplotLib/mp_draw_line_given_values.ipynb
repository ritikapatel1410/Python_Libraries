 @Title : draw a line using given axis values with suitable label in the x axis, y axis and a title

```python
import matplotlib.pyplot as plt 
import numpy as np
```

```python
x_axis_values=np.array([0,1,2,3])
y_axis_values=np.array([15,1,7,3])
```

```python
plt.plot(x_axis_values,y_axis_values)
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.title("Draw line")
```

```python
plt.show()
```
