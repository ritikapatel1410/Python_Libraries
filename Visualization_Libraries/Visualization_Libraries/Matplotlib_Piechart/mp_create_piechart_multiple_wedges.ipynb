 @Title : create a pie chart with a title of the popularity of programming Languages

```python
import matplotlib.pyplot as plt 
```

```python
languages = 'Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++'
popuratity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
colors = ["blue", "red", "green", "purple", "yellow", "pink"]
wedges = (0.1, 0, 0, 0, 0, .1)  
```

```python
plt.pie(popuratity,labels=languages, colors=colors,autopct='%1.1f%%',explode=wedges,shadow=True)
plt.title("Popularity of Programming Language", bbox={'facecolor':'white', 'pad':5})
plt.show()
```
