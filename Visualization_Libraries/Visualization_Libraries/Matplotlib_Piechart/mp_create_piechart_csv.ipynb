 @Title : create a pie chart of gold medal achievements of five most successful countries in 2016 Summer Olympics.Read the data from a csv file

```python
import matplotlib.pyplot as plt 
import pandas as pd
```

```python
dataframe=pd.read_csv('csv_file/medal.csv')
country_data = dataframe["country"]
medal_data = dataframe["gold_medal"]
colors = ["blue", "red", "green", "purple", "yellow", "pink"]
```

```python
plt.pie(medal_data,labels=country_data, colors=colors,autopct='%1.1f%%')
plt.show()
```
