```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic_dataset = sns.load_dataset("titanic")
sns.set_style("dark")
Barchart=sns.barplot(x = "sex", y = "survived", data = titanic_dataset)
Barchart.set_title("Titanic sex vs servived barchart",color="blue",fontsize=19)
plt.show()
```
