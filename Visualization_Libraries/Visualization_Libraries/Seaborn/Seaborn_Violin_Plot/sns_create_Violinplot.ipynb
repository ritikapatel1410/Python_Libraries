 @Title : draw a violin plot of sex against day for a given dataset

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips_dataset = sns.load_dataset("tips")
sns.set_style('darkgrid')
violinplot=sns.violinplot(x = "day", y = "total_bill",hue="sex",data = tips_dataset)
violinplot.set_title("Tips day vs total bill violin plot",color="blue",fontsize=19)
plt.show()
```
