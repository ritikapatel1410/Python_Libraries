 @Title : draw a barplot to show count for a deck for a dataset

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic_dataset = sns.load_dataset("titanic")
sns.set_style("dark")
#Barchart=sns.countplot(x = "deck",data = titanic_dataset)
Barchart=sns.barplot(x=titanic_dataset.deck.value_counts().index, y=titanic_dataset.deck.value_counts(),data =titanic_dataset)
Barchart.set_title("Titanic deck vs survived barchart",color="blue",fontsize=19)
plt.show()
```
