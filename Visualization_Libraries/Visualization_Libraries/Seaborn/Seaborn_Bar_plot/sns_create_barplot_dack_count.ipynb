 @Title : draw a barplot to show count for a deck for a dataset

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic_dataset = sns.load_dataset("titanic")
print(titanic_dataset.head())
sns.set_style("dark")
Barchart=sns.barplot(x = "deck", y = "survived", data = titanic_dataset)
Barchart.set_title("Titanic deck vs servived barchart",color="blue",fontsize=19)
plt.show()
```
