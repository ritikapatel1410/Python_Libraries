 @Title : draw a scatter plot of “day” against “total bill”

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips_dataset = sns.load_dataset("tips")
sns.set_style('darkgrid')
print(tips_dataset.head())
Scatterplot=sns.scatterplot(x = "day", y = "total_bill", data = tips_dataset,hue="sex")
Scatterplot.set_title("Tips day vs total bill Scatterplot",color="blue",fontsize=19)
plt.show()
```
