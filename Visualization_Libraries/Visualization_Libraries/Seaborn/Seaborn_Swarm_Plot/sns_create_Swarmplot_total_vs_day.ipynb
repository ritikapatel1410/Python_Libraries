 @Title : draw swarm plot of “total bill” against day

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips_dataset = sns.load_dataset("tips")
sns.set_style('whitegrid')
violinplot=sns.violinplot(x = "total_bill", y = "day",hue="sex",data = tips_dataset)
violinplot.set_title("draw swarm plot of “total bill” against day",color="blue",fontsize=19)
plt.show()
```
