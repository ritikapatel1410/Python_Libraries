 @Title : draw a swarm plot of total bill against size tips dataset

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips_dataset = sns.load_dataset("tips")
sns.set_style('whitegrid')
violinplot=sns.violinplot(x = "total_bill", y = "size",hue="sex",data = tips_dataset)
violinplot.set_title("swarm plot of total bill against size ",color="blue",fontsize=19)
plt.show()
```
