```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic_dataset = sns.load_dataset("titanic")
sns.set_style("dark")
Pointplot=sns.pointplot(x = "sex", y = "survived", data = titanic_dataset,linestyles=":")
Pointplot.set_title("Titanic gender vs servived Pointplot",color="blue",fontsize=19)
plt.show()
```
