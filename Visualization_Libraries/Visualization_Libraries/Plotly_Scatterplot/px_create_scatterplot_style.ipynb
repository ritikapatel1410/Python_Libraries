 @Title : Write a program to draw a scatter plot for random 500 x and y coordinates and style it

```python
import plotly.graph_objects as go
import random
import numpy as np
```

```python
x_values = [random.triangular() for i in range(50)]
y_values = [random.gauss(0.5, 0.25) for i in range(50)]
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=x_values, y=y_values,name="xy points",mode='markers',marker=dict(size=16,color=np.random.randn(500), #set color equal to a variable
        colorscale='Viridis', # one of plotly colorscales
        showscale=True,
        line_width=3
     )))
fig.update_layout(title='random distribution',yaxis_zeroline=False, xaxis_zeroline=False)
fig.show()
```
