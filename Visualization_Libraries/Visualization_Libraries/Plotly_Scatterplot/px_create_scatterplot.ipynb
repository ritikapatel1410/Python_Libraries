 @Title : draw a scatter plot for random 1000 x and y coordinates

```python
import plotly.express as px
import random
```

```python
x_values = [random.triangular() for i in range(1000)]
y_values = [random.gauss(0.5, 0.25) for i in range(1000)]
```

```python
fig = px.scatter(x=x_values, y=y_values,labels={'x':'x-axis', 'y':'y-axis'})
fig.update_layout(title='random distribution',yaxis_zeroline=False, xaxis_zeroline=False)
fig.show()
```
