 @Title : draw a scatter plot to find sea level rise in past 100 years

```python
import matplotlib.pyplot as plt
import pandas as pd 
```

```python
data = pd.read_csv('csv_file/data.csv')
year = data['Year']
sea_levels = data['CSIRO Adjusted Sea Level']
```

```python
plt.scatter(year, sea_levels, edgecolors='g')
plt.xlabel('year', fontsize=16)
plt.ylabel('height', fontsize=16)
plt.title('sea level scatter plot',fontsize=20)
plt.show()
```
