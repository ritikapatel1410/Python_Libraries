 @Title : draw a scatter plot with empty circles taking a random distribution in X and Y and plotted against each other.

```python
import matplotlib.pyplot as plt
from pylab import randn
```

```python
x_value = randn(200)
y_value = randn(200)
plt.scatter(x_value,y_value, facecolors='none', color='r',s=60)
plt.xlabel("x- axis")
plt.ylabel("y-axis")
plt.title("random distribution in X and Y using empty circle")
plt.show()
```
