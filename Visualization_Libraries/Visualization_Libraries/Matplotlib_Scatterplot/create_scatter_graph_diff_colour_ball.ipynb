 @Title : draw a scatter plot using random distributions to generate balls of different sizes other

```python
import matplotlib.pyplot as plt
import random
import math
```

```python
no_of_balls = 25
x_value = [random.triangular() for i in range(no_of_balls)]
y_value = [random.gauss(0.5, 0.25) for i in range(no_of_balls)]
colors = [random.randint(1, 4) for i in range(no_of_balls)]
areas = [math.pi * random.randint(5, 15)**2 for i in range(no_of_balls)]
plt.scatter(x_value, y_value, s=areas, c=colors, alpha=0.85)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("scatter plot with different colour balls")
plt.show()
```
