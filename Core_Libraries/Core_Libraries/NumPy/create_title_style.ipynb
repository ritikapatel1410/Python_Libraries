```python
import matplotlib.pyplot as plt
```

```python
languages = 'Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++'
popuratity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
colors = ["blue", "red", "yellow", "white", "purple", "pink"]
explode = (0.1, 0, 0, 0, 0, 0)
```

```python

a,b,autotexts=plt.pie(popuratity, explode=explode, labels=languages, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=140,pctdistance=0.6)
for autotext in autotexts:
    autotext.set_color('black')
    autotext.set_weight('bold')
    autotext.set_fontsize(9)
plt.title("PopularitY of Programming Language\n" + "Worldwide, Oct 2017 compared to a year ago", bbox={'facecolor':'white', 'pad':5},c="black",fontsize=16,weight="bold",family="serif")
plt.legend(loc="right",fontsize=12)
plt.show()
```
